```python
import os
from pathlib import Path
import pandas as pd
from dotenv import load_dotenv

# Load environment variables from .env file in project root
load_dotenv()

# Get CSV path from environment variable
csv_env = os.getenv("NAMUS_CSV_PATH")
if not csv_env:
    raise RuntimeError(
        "NAMUS_CSV_PATH is not set. Please create a .env file in the project root "
        "with: NAMUS_CSV_PATH=your/path/to/NamUs_Master_Robust_v3.csv"
    )

csv_path = Path(csv_env)

# Read the CSV into a pandas DataFrame
df = pd.read_csv(csv_path)

print(f"Using CSV path: {csv_path}")
print(f"DataFrame loaded with shape: {df.shape}")
```

```python
# Simple checks on the NamUs DataFrame `df`

print("DataFrame head:")
print(df.head())

print("\nDataFrame info:")
print(df.info())

print("\nDataFrame columns:")
print(df.columns)
```
